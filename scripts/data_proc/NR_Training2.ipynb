
## adj-pad (2-way)

```python
new_rows_list = []

for i in range(1, len(df) - 1):
    prev_row = df.iloc[i - 1, :]
    next_row = df.iloc[i + 1, :]
    row = df.iloc[i, :]

    prev_row = prev_row.add_prefix('prev_')
    next_row = next_row.add_prefix('next_')
    
    new_rows_list.append(pd.concat([prev_row, row, next_row]))


adj_2w_df = pd.concat(new_rows_list, axis=1).transpose()
```

```python
date_batches = [
    [['2001-07-11', '2001-07-12', '2001-07-13', '2001-07-14', '2001-07-15'], ['2001-07-13',]],
    [['2001-09-16', '2001-09-17', '2001-09-18', '2001-09-19', '2001-09-20'], ['2001-09-18',]],
    [['2003-01-23', '2003-01-24', '2003-01-25', '2003-01-26', '2003-01-27'], ['2003-01-25',]],
    [['2005-01-12', '2005-01-13', '2005-01-14', '2005-01-15', '2005-01-16', '2005-01-17', '2005-01-18', '2005-01-19'], ['2005-01-14', '2005-01-15', '2005-01-16', '2005-01-17',]],
    [['2014-03-31', '2014-04-01', '2014-04-02', '2014-04-03', '2014-04-04'], ['2014-04-02',]]
]
```

```python
date_adj_pad = dict()
for db_idx, date_batch in enumerate(date_batches):
    for idx, date in enumerate(date_batch[0]):
        if idx == 0:
            date_adj_pad[date] = (None, date_batch[0][idx+1])
        elif idx == len(date_batch[0]) - 1:
            date_adj_pad[date] = (date_batch[0][idx-1], None)
        else:
            date_adj_pad[date] = (date_batch[0][idx-1], date_batch[0][idx+1])
```

```python
date_adj_pad
```

### Calculating the average values of each date to use as prev/next day values

```python
def extract_ym(date: str):
    return date[:7]
```

```python
avg_df_rows = dict()

for date in date_adj_pad:
    filtered_df = df[df["date"] == date]
    avg_df_rows[date] = filtered_df.select_dtypes(include='number').mean()
```

```python
# Adding YYYY-MM avg (The values seem to increase with time so overall average may not be the best thing to do)
unique_ym = set()
for date in date_adj_pad:
    unique_ym.add(date[:7])
unique_ym

filtered_df = df[df["is_anamoly"] == 0.0]
for ym in unique_ym:
    _filtered_df = filtered_df[filtered_df['date'].str.startswith(ym)]
    avg_df_rows[ym] = filtered_df.select_dtypes(include='number').mean()
```

```python
# Adding the overall non-anomalous avg row

filtered_df = df[df["is_anamoly"] == 0.0]
avg_df_rows['non_anomalous_avg'] = filtered_df.select_dtypes(include='number').mean()
```

### Constructing the 1-day pad 2-way df

```python
oneD_2W_pad_rows = []

for idx, row in df.iterrows():
    curr_date = row['date']
    prev_date, next_date = date_adj_pad[curr_date]

    if prev_date is None:
        prev_row = avg_df_rows[extract_ym(curr_date)]
    else:
        prev_row = avg_df_rows[prev_date]
    prev_row = prev_row.add_prefix('prev_')
    
    if next_date is None:
        next_row = avg_df_rows[extract_ym(curr_date)]
    else:
        next_row = avg_df_rows[next_date]
    next_row = next_row.add_prefix("next_")
    
    oneD_2W_pad_rows.append(pd.concat([prev_row, row, next_row]))
```

```python
oneD_2W_pad_df = pd.DataFrame()
oneD_2W_df = pd.concat(oneD_2W_pad_rows, axis=1).transpose()
```
